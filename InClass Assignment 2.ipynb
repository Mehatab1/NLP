
Task: Create a small search or recommendation system that matches a user query with the most relevant text documents using TF-IDF and cosine similarity.

Example Scenario
You have 10 movie descriptions. A user searches: “space adventure with robots”
Your system should recommend the movie descriptions that are most similar to this query

Create a dataset
Use at least 10 short text documents
Each document should have a title and description
Preprocess the text
Lowercase
Remove punctuation
Remove stopwords
Lemmatize or stem words
Convert text into TF-IDF vectors
Use tf-idf vectorizer from sklearn
Accept a user query
Example: "healthy food"
Preprocess the query in the same way as the documents
Calculate cosine similarity
Compare the query vector with all document vectors
Return the top 3 results
Show title
Show description
Show similarity score

```python
import pandas as pd
import string
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
movies = [
    {
        "title": "Galaxy Warriors",
        "description": "A space adventure where robots help humans fight alien enemies."
    },
    {
        "title": "Healthy Kitchen",
        "description": "A guide to healthy food recipes with fresh vegetables and fruits."
    },
    {
        "title": "Ocean Mystery",
        "description": "Divers discover hidden treasures deep under the sea."
    },
    {
        "title": "Robot Future",
        "description": "Advanced robots and artificial intelligence change human life."
    },
    {
        "title": "Fitness First",
        "description": "Tips for exercise healthy eating and maintaining fitness."
    },
    {
        "title": "Space Journey",
        "description": "Astronauts travel through space on a dangerous mission."
    },
    {
        "title": "Cooking Master",
        "description": "Delicious homemade food and easy cooking techniques."
    },
    {
        "title": "Alien Planet",
        "description": "Explorers land on a mysterious alien world full of danger."
    },
    {
        "title": "Tech Revolution",
        "description": "Innovative technology and smart machines shape the future."
    },
    {
        "title": "Nature Escape",
        "description": "A peaceful journey through forests mountains and rivers."
    }
]

df = pd.DataFrame(movies)
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))

    words = text.split()

    filtered_words = []

    for word in words:
        if word not in stop_words:
            filtered_words.append(lemmatizer.lemmatize(word))

    return " ".join(filtered_words)
```

```python
df["clean_description"] = df["description"].apply(clean_text)

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df["clean_description"])

query = input("Enter your search query: ")

clean_query = clean_text(query)

query_vector = vectorizer.transform([clean_query])

scores = cosine_similarity(query_vector, tfidf_matrix)

df["score"] = scores[0]

top_matches = df.sort_values(by="score", ascending=False).head(3)

print("\nTop Recommendations:\n")

for _, row in top_matches.iterrows():
    print("Title:", row["title"])
    print("Description:", row["description"])
    print("Similarity Score:", round(row["score"], 4))
    print()
```
