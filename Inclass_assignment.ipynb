```python
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

sentence = "I saw the man with a telescope."
doc = nlp(sentence)

print("Token\tDependency\tHead")

for token in doc:
    print(f"{token.text}\t{token.dep_}\t\t{token.head.text}")

# Find ROOT
for token in doc:
    if token.dep_ == "ROOT":
        print("\nROOT WORD:", token.text)
```

```python
print("""

1. The ROOT word is saw because it is the main verb.

2. The phrase with a telescope modifies man  according to spaCy.

3. The sentence is ambiguous because:
   - I used a telescope to see the man.
   - The man had a telescope.

4. spaCy chose the interpretation where the man has the telescope.

5. Rewritten sentences:
   -I used a telescope to see the man.
   -I saw the man who was carrying a telescope.""")
```

```python
import spacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")

text = '''
Microsoft announced a new AI research initiative in Berlin on 15 June 2026.
The project involves collaboration with Google, SAP, and SRH University Heidelberg.
According to Satya Nadella, the initiative will expand AI adoption across Germany.
'''

doc = nlp(text)

print("Detected Entities:\n")

for ent in doc.ents:
    print(ent.text, "-->", ent.label_)

# Count entity labels
labels = [ent.label_ for ent in doc.ents]
counter = Counter(labels)

print("\nEntity Counts:")
print(counter)
```

### Answers

1. Detected entities:
   - Microsoft
   - Berlin
   - 15 June 2026
   - Google
   - SAP
   - SRH University Heidelberg
   - Satya Nadella
   - Germany

2. Entity labels:
   - Microsoft → ORG
   - Berlin → GPE
   - 15 June 2026 → DATE
   - Google → ORG
   - SAP → ORG
   - SRH University Heidelberg → ORG
   - Satya Nadella → PERSON
   - Germany → GPE

3. The most common entity type is **ORG**.

4. spaCy identified most entities correctly.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "Artificial intelligence is transforming healthcare through predictive analytics.",

    "Machine learning and AI are improving disease diagnosis in hospitals.",

    "Football teams are preparing for the international championship tournament."
]

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

# Cosine Similarity
similarity_matrix = cosine_similarity(tfidf_matrix)

print("Cosine Similarity Matrix:\n")
print(similarity_matrix)

print("\nSimilarity between Document 1 and 2:",
      similarity_matrix[0][1])

print("Similarity between Document 1 and 3:",
      similarity_matrix[0][2])

print("Similarity between Document 2 and 3:",
      similarity_matrix[1][2])
```

### Answers

1. Documents 1 and 2 are the most similar because both discuss AI and healthcare.

2. Yes, the result is reasonable because both documents belong to similar domains.

3. TF-IDF gives higher importance to words that are frequent in one document but rare across all documents.

4. Cosine similarity does not truly understand meaning. It only compares mathematical vectors.

5. TF-IDF does not automatically understand that "AI" and "machine learning" are related concepts.
